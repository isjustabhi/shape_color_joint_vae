```python
# view_results.ipynb (Python version for copy-paste into notebook)

import os
from PIL import Image
import matplotlib.pyplot as plt

# Folder where outputs are stored
samples_dir = "samples/joint"

# Extract available epoch numbers
epochs = sorted([
    int(f.split("epoch")[1].split(".")[0])
    for f in os.listdir(samples_dir)
    if f.startswith("input_epoch") and f.endswith(".png")
])

# Show side-by-side input and reconstruction
def show_epoch(epoch):
    input_path = os.path.join(samples_dir, f"input_epoch{epoch}.png")
    output_path = os.path.join(samples_dir, f"output_epoch{epoch}.png")
    
    if os.path.exists(input_path) and os.path.exists(output_path):
        input_img = Image.open(input_path)
        output_img = Image.open(output_path)

        plt.figure(figsize=(10, 5))

        plt.subplot(1, 2, 1)
        plt.title(f"Input (Epoch {epoch})")
        plt.imshow(input_img)
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.title("Reconstruction")
        plt.imshow(output_img)
        plt.axis('off')

        plt.tight_layout()
        plt.show()
    else:
        print(f"Missing files for epoch {epoch}.")

# Show latest
#ow_epoch(epochs[-1])

# Or preview multiple:
for e in epochs:
     show_epoch(e)
```
